```python
import openpyxl
import pandas as pd
import os
import datetime
import copy
```

```python
excelname='raw_data'
sheetname='data'
wb=openpyxl.load_workbook(str(excelname)+'.xlsx')
ws=wb[str(sheetname)]
```

```python
dataset=[]
i=1
while(ws.cell(row=i,column=1).value!=None):
    row=[]
    j=1
    while(ws.cell(row=i,column=j).value!=None):
        row.append(ws.cell(row=i,column=j).value)
        j+=1
    dataset.append(row)
    i+=1
```

```python
#（3*c(t)＋2*c(t-1)＋1*c(t-2)）/ 6 ，值高於收盤價 買 ，值低於收盤價  賣
```

```python
trade = []
for i in range(0, len(dataset)):
    row = []
    for j in range(0, 14):
        row.append(0)
    trade.append(row)
pd.DataFrame(trade)
```

```python
trade[0][0] = "date"
trade[0][1] = "open"
trade[0][2] = "high"
trade[0][3] = "low"
trade[0][4] = "close"
trade[0][5] = "wma"
trade[0][6] = "buy"
trade[0][7] = "hold"
trade[0][8] = "sell"
trade[0][9] = "number"
trade[0][10] = "value"
trade[0][11] = "period_return"
trade[0][12] = "holding_period"
trade[0][13] = "annual return"
```

```python
for i in range(1, len(trade)):
    trade[i][0] = dataset[i][0]
    trade[i][1] = float(dataset[i][1])
    trade[i][2] = float(dataset[i][2])
    trade[i][3] = float(dataset[i][3])
    trade[i][4] = float(dataset[i][4])
pd.DataFrame(trade).head()
```

```python
for i in range(3, len(trade)):
    #temp = 0
    #for j in range(0, 3):
        #temp = temp + trade[i-j][4]
    trade[i][5] = (trade[i][4]*3+trade[i-1][4]*2+trade[i-2][4]*1)/6
    
pd.DataFrame(trade)
```

```python
#wma=（3*c(t)＋2*c(t-1)＋1*c(t-2)）/ 6
```

```python
for i in range(3, len(trade)):
    if trade[i][5]>trade[i][4]  and trade[i-1][7]!=1:
        trade[i][6]=1
    if trade[i][5]<trade[i][4] and trade[i-1][7]==1:
        trade[i][8]=1
    if trade[i][6]==1 or trade[i-1][7]==1:
        trade[i][7]=1
    if trade[i][8]==1:
        trade[i][7]=0
pd.DataFrame(trade)
```

```python
initial=10000
for i in range(3,len(trade)):
    if(trade[i][6]==1)and(trade[i][7]==1)and(trade[i][8]==0):
        trade[i][9]=initial/trade[i][4]
        trade[i][10]=trade[i][9]*trade[i][4]
        trade[i][11]=round((trade[i][9]*trade[i][4]-initial)/initial,4)
    elif(trade[i][6]==0)and(trade[i][7]==0)and(trade[i][8]==1):
        trade[i][9]=0
        trade[i][10]=trade[i-1][9]*trade[i][4]
        trade[i][11]=round((trade[i-1][9]*trade[i][4]-initial)/initial,4)
    elif(trade[i][6]==0)and(trade[i][7]==1)and(trade[i][8]==0):
        trade[i][9]=trade[i-1][9]
        trade[i][10]=trade[i][9]*trade[i][4]
        trade[i][11]=round((trade[i][9]*trade[i][4]-initial)/initial,4)
pd.DataFrame(trade)
```

```python
import datetime
from datetime import date
for i in range(1,len(trade)):
    trade[i][0]=datetime.datetime.strptime(trade[i][0],'%Y/%m/%d')
```

```python
first=0
end=0
for i in range(1,len(trade)):
    if(trade[i][6]==1):
        trade[i][12]=1
        first=trade[i][0]
    elif((trade[i][7]>0)or(trade[i][8]==1)and(trade[i][6]!=1)):
        end=trade[i][0]
        trade[i][12]=(end-first).days+1
pd.DataFrame(trade)
```

```python
for i in range(1,len(trade)):
    if(trade[i][12]!=0):
        trade[i][13]=(1+trade[i][11])**(365/trade[i][12])-1
```

```python
annual_return=1
trading=0
for i in range(1,len(trade)):
    if(trade[i][8]==1):
        annual_return*=(1+trade[i][11])
        trading+=1

final_return=annual_return**(1/trading)-1
```
